# ***Get DATAFRAME ('Tableau') from file using pandas where file is csv format***

```python
import pandas as pd
data = pd.read_csv('DragonMSI-IC50-All.txt',delimiter='\t')
```

```python
print(data)
```

# ***Show the loaded data in table format***

```python
data.head()
```

# ***Load using .iloc[rows,columns] the targeted columns and rows, and convert it to vectors using .values***

```python
x = data.iloc[:,9:]
```

```python
print(x)
```

```python
y = data.iloc[:,2]
```

```python
print(y)
```

# ***Data processing***

# **Normalization **

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler object
scaler = StandardScaler()

# Transform the data
x = scaler.fit_transform(x)
```

# ***Construire l'ensemble de teste et l'ensemble d'apprentissage***

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
```

# ***Training the model***

```python
from sklearn.ensemble import BaggingRegressor
model = BaggingRegressor()
model.fit(X_train,y_train)
```

```python
print(X_test.shape)
```

# ***Evaluation du modèle en utilisant la partie Teste***

```python
from sklearn.metrics import *
y_pred = model.predict(X_test)
```

```python
print(y_pred.shape)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
print(mae,mse,r2)
```

# ***Utiliser une technique de validation croisé***

```python
model = BaggingRegressor()
from sklearn.model_selection import cross_validate
scores = cross_validate(model,x, y, cv=5,scoring=['r2','neg_mean_absolute_error','neg_mean_squared_error'])
print(scores)
```
